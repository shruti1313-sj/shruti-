```python
# Q1: join tuples if similar initial element

# ANS

l=[(5,6),(5,7),(5,8),(6,10),(7,13)]
print("the original list is : "+str(l))
s=[]
for i in l:
  if s and s[-1][0]==i[0]:
     s[-1].extend(i[1:])
  else:
    s.append([ele for ele in i])
s = list(map(tuple, s))
print("the extracted element:"+str(s))
```

```python
# Q2 : multiply adjacent element (both side) and take sum of right and left side multipication result

# ANS

tup=(1,5,7,8,10)
print("the original tuple :")
print(tup)
res=tuple(i*j for i,j in zip(tup,tup[1:]))
print("the result is :")
print(res)
```

```python
# Q5: intersection of two lists. intersection of two list means we need to take
#     all those element which are common to both of the initial list and store them into another list .
#     only use  list comprehension

# ANS


def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3

lst1 = [4, 9, 1, 17, 11, 26, 28, 54, 69]
lst2 = [9, 9, 74, 21, 45, 11, 63, 28, 26]
print(intersection(lst1, lst2))
```

```python
# Q1 : key with maximum unique value

# ANS

test_dict = {"campusX": [5, 7, 9, 4, 0],
             "is": [6, 7, 4, 3, 3],
             "Best": [9, 9, 6, 5, 5]}

print("The original dic: " )
print(test_dict)

max_val = 0
max_key = None

for sub in test_dict:


    if len(set(test_dict[sub])) > max_val:
        max_val = len(set(test_dict[sub]))
        max_key = sub
print("Key with maximum unique values : " + str(max_key))
```

```python
# Q2: replace words from dictionary. Given string , replace its words from lookup dictionary

# ANS

test_str = "campusx best for DS students."


print(test_str)

repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}

temp = test_str.split()
res = []
for wrd in temp:

    res.append(repl_dict.get(wrd, wrd))

res = ' '.join(res)


print("Replaced Strings : " + str(res))
```

```python
# Q3: convert list to list of dictionaries. Given list values and keys list,
#      convert these values to keys list , convert these values to key value pairs in from of list of dictionaries

# ANS

d1=["Data science", 3,"is",8]
print("the original list is :")
print(d1)
d2 = ["name", "id"]


n = len(d1)
res = []
for idx in range(0, n, 2):
    res.append({d2[0]: d1[idx], d2[1] : d1[idx + 1]})

print("The constructed dictionary list : " )
print(res)
```

```python
# Q4 convert a list of tuples into dictionary

# ANS

def Convert(tup, di):
    for a, b in tup:
        di.setdefault(a, []).append(b)
    return di


tups = [("shruti", 10), ("rutuja", 12), ("anandi", 14),
        ("aaditi", 20), ("seeta", 25), ("geeta", 30)]
dictionary = {}
print(Convert(tups, dictionary))
```

```python
# Q5: sort dictionary key and values list

# ANS

a1 = {'c':[3],'b':[12,10],'a':[19,4]}
print("The original dictionary is : " )
print(a1)


res = {key : sorted(a1[key]) for key in sorted(a1)}


print("The sorted dictionary : ")
print(res)
```
